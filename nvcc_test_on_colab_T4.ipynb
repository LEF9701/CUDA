```python
%%writefile hello.cu
#include <iostream>
#include <cuda_runtime.h>

__global__ void helloFromGPU() {
    printf("Hello World from GPU thread %d!\n", threadIdx.x);
}

int main() {
    printf("Hello World from CPU!\n");
    helloFromGPU<<<1, 5>>>();
    cudaDeviceSynchronize();
    return 0;
}
```

```python
!nvcc hello.cu -o hello
```
